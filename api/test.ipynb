```python
import folium

m = folium.Map(location=(40.109778, -88.228446), zoom_start=24, zoom_control=False)
```

```python
import pandas as pd
df = pd.read_csv('sites.csv')
```

```python
df.head()
```

```python
def show_site(site_id):
    row = df[df['Site ID']==site_id]
    folium.Marker(location=(row['Latitude'], row['Longitude'])).add_to(m)
    print(type(row))
    return row

show_site(2378)
```
